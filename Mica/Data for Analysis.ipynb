```python
# Read the CSV file
TH_State= pd.read_csv('TMEDTREND_PUBLIC_230821 FINAL.csv')

# Display the first few rows of the DataFrame
TH_State.head()
```

```python
# Filter rows with '2023' in the 'Year' column
TH_State_2023 = TH_State[TH_State['Year'] == 2023]
TH_State_2023.head()
```

```python
#Filter dataframe with State totals for all columns: 

TH_State_reduce=TH_State_2023[
    (TH_State_2023['Bene_Mdcd_Mdcr_Enrl_Stus'] == "All") &
    (TH_State_2023['Bene_Race_Desc'] == "All") &
    (TH_State_2023['Bene_Sex_Desc'] == "All") &
    (TH_State_2023['Bene_Mdcr_Entlmt_Stus'] == "All") &
    (TH_State_2023['Bene_Age_Desc'] == "All") &
    (TH_State_2023['Bene_RUCA_Desc'] == "All")
]
TH_State_reduce.head()
```

```python
# Dropping rows with Rows not included in 50 US States definition in certain columns

TH_State_reduced = TH_State_reduce[
    (TH_State_reduce['Bene_Geo_Desc'] != "District Of Columbia") &
    (TH_State_reduce['Bene_Geo_Desc'] != "Missing Data") &
    (TH_State_reduce['Bene_Geo_Desc'] != "National") &
    (TH_State_reduce['Bene_Geo_Desc'] != "Puerto Rico") &
    (TH_State_reduce['Bene_Geo_Desc'] != "Territories") &
    (TH_State_reduce['Bene_Geo_Desc'] != "Virgin Islands")
]
TH_State_reduced.head()
```

```python
# DataFrame with State & All Pct_Telehealth
TH_State_AllTH = TH_State_reduced[["Year", "Bene_Geo_Desc", "Pct_Telehealth"]]

# Rename Column Pct_Telehealth
TH_State_AllTH = TH_State_AllTH.rename(columns={"Pct_Telehealth": "State_Pct_Telehealth"})

# Display the first few rows of the DataFrame
TH_State_AllTH.head()
```

```python
# Filter dataframe for specific values in the 'Bene_RUCA_Desc' column
desired_values = ["Rural", "Urban"]
TH_State_filtered = TH_State_2023[TH_State_2023['Bene_RUCA_Desc'].isin(desired_values)]

# DataFrame with State, specified RUCA values, and Pct_Telehealth
TH_State_filtered = TH_State_filtered[["Year", "Bene_Geo_Desc", "Bene_RUCA_Desc", "Pct_Telehealth"]]

# Dropping rows with Rows not included in 50 US States definition in certain columns
TH_State_filtered = TH_State_filtered[
    (TH_State_filtered['Bene_Geo_Desc'] != "District Of Columbia") &
    (TH_State_filtered['Bene_Geo_Desc'] != "Missing Data") &
    (TH_State_filtered['Bene_Geo_Desc'] != "National") &
    (TH_State_filtered['Bene_Geo_Desc'] != "Puerto Rico") &
    (TH_State_filtered['Bene_Geo_Desc'] != "Territories") &
    (TH_State_filtered['Bene_Geo_Desc'] != "Virgin Islands")
]

TH_State_filtered.head(50)
```

```python
# Creating a pivot table
pivot_table = TH_State_filtered.pivot_table(values='Pct_Telehealth', index='Bene_Geo_Desc', columns='Bene_RUCA_Desc')

# Rename Rural and Urban column names
pivot_table = pivot_table.rename(columns={"Rural": "Rural_Pct_Telehealth", "Urban": "Urban_Pct_Telehealth"})

# Displaying the pivot table
print(pivot_table)
```

```python
#Merge TH_State_AllTH with pivot_table

merge_TH_data=pd.merge(TH_State_AllTH, pivot_table, on="Bene_Geo_Desc")

#Rename Bene_Geo_Desc column to "state"
merge_TH_data=merge_TH_data.rename(columns={"Bene_Geo_Desc": "state"})

merge_TH_data.head()
```

```python
#Merge TH with C2H data

analysis_data=pd.merge(merge_TH_data, C2H_data_reduce, on="state")
analysis_data.head()
```
